### Use the artist layer of Matplotlib to replicate the bar chart below to visualize the percentage of the respondents' interest in the different data science topics surveyed.

```python
%matplotlib inline 
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')

df = pd.read_csv('C:/Users/lucas/Documents/lucastiagooliveira/Study Materials/IBM Couse - Data Science/Data Visualization with Python/Topic_Survey_Assignment.csv')
df.rename(columns = {'Unnamed: 0':'Topics'}, inplace = True)
df.set_index('Topics', inplace = True)
df
```

```python
df = df.sort_values(by='Very interested', ascending = False)

df.set_index('Topics', inplace = True)
total = df[['Very interested','Somewhat interested', 'Not interested']].sum(axis=0).sum()
```

```python
df.iloc[1][0]/df.iloc[:,0].sum()
```

```python
color_bar = ['#5cb85c', '#5bc0de' ,'#d9534f']
ax = df.plot(kind = 'bar', figsize = (20,8),
        width = 0.8,
        color = color_bar
       )
plt.legend(loc = 'upper right', fontsize = 14, facecolor = 'white')
plt.title("Percentage of Respondents' Interest in Data Science Areas", fontsize = 16)
plt.xlabel('')


ax.patch.set_facecolor('white')

ax.spines['bottom'].set_color('black')

ax.axes.get_yaxis().set_visible(False)

for i in range(0,df.shape[0]):
    plt.annotate(str(round((df.iloc[i][0]/df.iloc[:,0].sum())*100,2)) + '%',
                 xy = (i - 0.35, df.iloc[i][0] + 25) ,
                 fontsize = 10)
    plt.annotate(str(round((df.iloc[i][1]/df.iloc[:,1].sum())*100,2)) + '%',
                 xy = (i - 0.08, df.iloc[i][1] + 25),
                 fontsize = 10)
    plt.annotate(str(round((df.iloc[i][2]/df.iloc[:,2].sum())*100,2)) + '%',
                 xy = (i + 0.2, df.iloc[i][2] + 25),
                 fontsize = 10)


plt.show()
```
