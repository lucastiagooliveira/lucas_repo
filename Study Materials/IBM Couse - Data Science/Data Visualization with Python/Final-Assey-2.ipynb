```python
df_new = df.groupby('PdDistrict').count()
df_new = df_new[['IncidntNum']]
df_new.reset_index(inplace = True)
df_new = df_new.rename(columns = {'PdDistrict':'Neighborhood','IncidntNum':'Count'})
df_new
```

```python
# San Francisco latitude and longitude values
latitude = 37.77
longitude = -122.42

san_geo = (r'san-francisco.json')
# len(san_geo)
# for feature in san_geo:
#     print(feature.properties['DISTRICT'])

sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12, tiles='Mapbox Bright')

# generate choropleth map using the total immigration of each country to Canada from 1980 to 2013
sanfran_map.choropleth(
    geo_data=san_geo,
    data=df_new,
    columns=['Neighborhood', 'Count'],
    key_on='root.crs.properties.name',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crimes in San Francisco'
)
```
